# Getting Started with cognitive3dpy

This notebook walks through the main data streams available in `cognitive3dpy`:
sessions, events, objectives, session-level objective steps, and exit polls.

## Installation

```bash
pip install cognitive3dpy
```

For pandas output support:
```bash
pip install "cognitive3dpy[pandas]"
```

## Authentication

Store your API key in a `.env` file (never commit this):

```
C3D_API_KEY=your_api_key_here
```

Then authenticate and set your project:

````python
from dotenv import load_dotenv
load_dotenv()  # loads C3D_API_KEY from .env

import cognitive3dpy as c3d

c3d.c3d_auth()       # reads C3D_API_KEY from environment
c3d.c3d_project(2797)   # replace with your project ID
````

## Sessions

`c3d_sessions()` returns one row per session with device info, location, and Cognitive3D metric scores.

By default it returns the last 30 days and ~40 curated columns (`compact=True`).

````python
sessions = c3d.c3d_sessions(n=200)
print(sessions.shape)
sessions.head()
````

````python
# Custom date range
sessions = c3d.c3d_sessions(
    start_date="2025-10-01",
    end_date="Mar 1, 2026"
)
sessions.sort(['session_date']).head()
````

````python
# All columns (not just compact)
sessions_full = c3d.c3d_sessions(n=100, compact=False)
print(sessions_full.columns)
````

### Scene sessions

Use `session_type="scene"` to get one row per session-scene combination.
Useful for comparing metrics across scenes within the same session.

````python
scene_sessions = c3d.c3d_sessions(session_type="scene", n=200)
scene_sessions.head()
````

````python
# Filter to a specific scene
scene_session_filtered = c3d.c3d_sessions(
    session_type="scene",
    scene_id="bd07b0ee-41d6-438d-a01c-ba0e3d92db44", # replace this with your scene ID
    n=200,
)
scene_session_filtered.head()
````

### Working with metric columns

All `c3d_metrics_*` and `c3d_metric_components_*` columns are `Float64`.

````python
import polars as pl

# Summary stats for top-level metric scores
metric_cols = [c for c in sessions.columns if c.startswith("c3d_metrics")]
sessions.select(metric_cols).describe()
````

````python
# Average FPS score by HMD
(
    sessions
    .group_by("c3d_device_model")
    .agg(pl.col("c3d_metrics_fps_score").mean().alias("avg_fps_score"))
    .sort("avg_fps_score", descending=True)
)
````

### Pandas output

Pass `output="pandas"` to any function to get a pandas DataFrame instead.

````python
sessions_pd = c3d.c3d_sessions(n=100, output="pandas")
sessions_pd.dtypes
````

## Events

`c3d_events()` returns one row per in-session event. Object IDs are resolved to friendly names and scene version IDs to scene names. Custom event properties appear as `prop_*` columns.

````python
events = c3d.c3d_events(start_date="2025-01-01", n=50)
print(events.shape)
events.head()
````

````python
# Most common event types
events.group_by("event_name").len().sort("len", descending=True)
````

````python
# Event positions — ready for spatial analysis
events.select(["event_name", "position_x", "position_y", "position_z", "object", "scene_name"])
````

## Objective Results

`c3d_objective_results()` returns aggregated success/failure counts across sessions.
Use `group_by="version"` (default) for high-level counts, or `group_by="steps"` for step-level detail.

````python
# Completion rates by objective version
results = c3d.c3d_objective_results(group_by="version")
results
````

````python
# Step-level breakdown
steps = c3d.c3d_objective_results(group_by="steps")
steps
````

````python
# Steps with low completion rates
steps.filter(pl.col("step_completion_rate") < 0.5).sort("step_completion_rate")
````

## Session Objectives

`c3d_session_objectives()` returns one row per objective step per session — useful for individual-level analysis.

````python
session_steps = c3d.c3d_session_objectives(start_date="2025-01-01")
session_steps.head()
````

````python
# Average step duration per step
(
    session_steps
    .group_by(["objective_name", "step_number", "step_description"])
    .agg(pl.col("step_duration_sec").mean().alias("avg_duration_sec"))
    .sort(["objective_name", "step_number"])
)
````

## Exit Polls

`c3d_exitpoll()` returns response counts per answer option, per question, per hook version.

````python
polls = c3d.c3d_exitpoll()
polls
````

````python
# Filter to a specific hook
polls = c3d.c3d_exitpoll(hook="end_questions_short", version=1)
polls
````

````python
# Response distribution for each question
(
    polls
    .with_columns(
        (pl.col("count") / pl.col("count").sum().over("question_title")).alias("pct")
    )
    .select(["question_title", "value_label", "count", "pct"])
    .sort(["question_title", "pct"], descending=[False, True])
)
````
